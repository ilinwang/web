### 計算所得稅
請寫出一個程式來計算使用者的年度綜合所得稅。這個程式會要求使用者輸入個人資訊 (姓名、電子信箱與所得淨額)，然後印出一段簡短的文字來告訴使用者應該繳交多少稅金。  
台灣的綜合所得稅採累進稅率，計算公式如下:  
>　　　　　所得淨額　　　　&nbsp;稅率  
> 　　　　0&nbsp;&nbsp;-&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;540,000　　　5%  
> 　540001&nbsp;&nbsp;-&nbsp;&nbsp;&nbsp;&nbsp;1,210,000　　&nbsp;12%  
> 1,210,001&nbsp;&nbsp;-&nbsp;&nbsp;&nbsp;&nbsp;2,420,000　　&nbsp;20%    
> 2,420,001&nbsp;&nbsp;-&nbsp;&nbsp;&nbsp;&nbsp;4,530,000　　&nbsp;30%  
> 4,530,001&nbsp;&nbsp;-&nbsp;&nbsp;&nbsp;10,310,000　&nbsp;&nbsp;&nbsp;40%  

舉例來說，假設某人的 $2018$ 年所得淨額為 $150$ 萬元，則  
1. 所得超過 $54$ 萬元，所以 $150$ 萬元中的 $54$ 萬元要被課第一區間的稅率 $5\%$
2. 又因為所得超過第二個區間的上限，所以 $150$ 萬元中要有 $121-54=67$ 萬元被課第二區間的稅率 $12\%$  
3. 所得沒有超過第三區間的上限，所以剩下的 $150-54-67=29$ 萬元要被課第三區間的稅率 $20\%$  
  
總計需繳稅額為 $54\times5\%+67\times12\%+29\times20\%=16.54$ (萬元)  
  
輸出的排版格式如下 (中英皆可):  
> 這個程式將幫您計算 2018 年度的綜合所得稅  
> 請輸入姓名: 王大明  
> 請輸入電子信箱: wangdaming@hahaha.net  
> 請輸入所得淨額: 1500000  
>   
> 親愛的王大明:  
> 您在 2018 年度的綜合所得稅為 16.54 萬元  
> 詳細的報告將在稍後為您寄至 wangdaming@hahaha.net

### 計算學期 GPA
GPA (grade point average) 是一種成績計算方式，常見於歐美，與等第制 (A+, A, A-, B+...) 成績可以互相轉換。台灣常用的成績計算為百分點制，若要換算單科成績為 GPA，可依下列方式:
> 80分以上: 4.0  
> 70 ~ 79 分: 3.0  
> 60 ~ 69 分: 2.0  
> 59 分以下: 1.0  

但若要計算總和 GPA，則必須考慮該科的學分數。舉例來說，假設某生成績與學分如下:
> 　　科目　　　百分制成績　　學分數  
> 　製造程序　　　　77　　　　　3  
> 　商用英文　　　　83　　　　　2  
> 　工程數學　　　　62　　　　　3  
> 　最佳化理論　　　91　　　　　3  

則該生的總和 GPA 為 $(3.0 \times 3+4.0 \times 2 + 2.0 \times 3 + 4.0 \times 3) \div (3+2+3+3)$，大概是 $3.18$  
  
請寫出一個程式，首先要求使用者輸入該學期修課的數量。接著，寫出一個迴圈，每進入一次迴圈都會讓使用者輸入課程名稱、期末成績、學分數等三樣資訊。最後，印出一句簡短的總結，告訴使用者這學期的 GPA 是多少 (四捨五入到小數點後第二位)。
